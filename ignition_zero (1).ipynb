# 🚀 Ignition Zero — O Início da Aurora

## Etapa 3: Script de Verificação de Telemetria

Este notebook lê os dados de telemetria da nave, verifica cada parâmetro contra as faixas seguras e determina se a decolagem está liberada ou deve ser abortada.

## 📋 Célula 1 — Dados de Telemetria

Insira aqui os valores reais da sua missão.

```python
# ============================================================
# DADOS DE TELEMETRIA — edite os valores abaixo
# ============================================================

telemetria = {

    # 🌡 Temperaturas (em °C)
    "temp_cabine":      22.5,   # seguro: 18 a 27 °C
    "temp_casco":       15.0,   # seguro: -50 a +60 °C
    "temp_motores":     35.0,   # seguro: 15 a 40 °C

    # 🏗 Integridade estrutural (1 = OK, 0 = FALHA)
    "integridade_estrutural": 1,

    # ⚡ Energia (em %)
    "bateria_principal": 97.0,  # seguro: 95–100% (mín. 85%)
    "bateria_backup":    90.0,  # seguro: 85–100%
    "energia_solar":     92.0,  # seguro: >= 90%

    # 🔧 Pressão dos tanques (em kPa)
    "pressao_combustivel": 325.0,  # seguro: 300–350 kPa
    "pressao_lox":         300.0,  # seguro: 280–320 kPa
    "pressao_helio_n2":   4200.0,  # seguro: 4.000–4.500 kPa
    "pressao_cabine":      101.0,  # seguro: 96–104 kPa

    # 🖥 Módulos críticos (1 = ONLINE, 0 = OFFLINE)
    "modulo_navegacao":      1,
    "modulo_controle_voo":   1,
    "modulo_comunicacao":    1,
    "modulo_suporte_vida":   1,
    "modulo_motores":        1,
    "modulo_abortagem":      1,
}

print("✅ Dados de telemetria carregados com sucesso.")
```

## 🔍 Célula 2 — Algoritmo de Verificação

```python
# ============================================================
# ALGORITMO DE VERIFICAÇÃO
# ============================================================

alertas = []   # lista de problemas encontrados
aprovado = True

def checar(condicao, mensagem):
    """Registra um alerta e marca decolagem como abortada se condição falhar."""
    global aprovado
    if not condicao:
        alertas.append(f"  ❌ {mensagem}")
        aprovado = False
    else:
        alertas.append(f"  ✅ {mensagem}")

print("=" * 50)
print("  VERIFICAÇÃO DE TELEMETRIA — IGNITION ZERO")
print("=" * 50)

# --- Temperatura ---
print("\n🌡 TEMPERATURA")
checar(18 <= telemetria["temp_cabine"] <= 27,
       f"Cabine: {telemetria['temp_cabine']}°C (faixa: 18–27°C)")
checar(-50 <= telemetria["temp_casco"] <= 60,
       f"Casco externo: {telemetria['temp_casco']}°C (faixa: -50 a +60°C)")
checar(15 <= telemetria["temp_motores"] <= 40,
       f"Motores: {telemetria['temp_motores']}°C (faixa: 15–40°C)")
for a in alertas[-3:]: print(a)

# --- Integridade ---
print("\n🏗 INTEGRIDADE ESTRUTURAL")
n = len(alertas)
checar(telemetria["integridade_estrutural"] == 1,
       f"Integridade: {telemetria['integridade_estrutural']} (esperado: 1)")
for a in alertas[n:]: print(a)

# --- Energia ---
print("\n⚡ ENERGIA")
n = len(alertas)
checar(telemetria["bateria_principal"] >= 85,
       f"Bateria principal: {telemetria['bateria_principal']}% (mín: 85%)")
checar(telemetria["bateria_backup"] >= 85,
       f"Bateria backup: {telemetria['bateria_backup']}% (mín: 85%)")
checar(telemetria["energia_solar"] >= 90,
       f"Energia solar: {telemetria['energia_solar']}% (mín: 90%)")
for a in alertas[n:]: print(a)

# --- Pressão ---
print("\n🔧 PRESSÃO DOS TANQUES")
n = len(alertas)
checar(300 <= telemetria["pressao_combustivel"] <= 350,
       f"Combustível: {telemetria['pressao_combustivel']} kPa (300–350)")
checar(280 <= telemetria["pressao_lox"] <= 320,
       f"LOX (oxidante): {telemetria['pressao_lox']} kPa (280–320)")
checar(4000 <= telemetria["pressao_helio_n2"] <= 4500,
       f"Hélio/N2: {telemetria['pressao_helio_n2']} kPa (4.000–4.500)")
checar(96 <= telemetria["pressao_cabine"] <= 104,
       f"Cabine: {telemetria['pressao_cabine']} kPa (96–104)")
for a in alertas[n:]: print(a)

# --- Módulos ---
print("\n🖥 MÓDULOS CRÍTICOS")
n = len(alertas)
modulos = {
    "modulo_navegacao":    "Navegação",
    "modulo_controle_voo": "Controle de voo",
    "modulo_comunicacao":  "Comunicação",
    "modulo_suporte_vida": "Suporte à vida",
    "modulo_motores":      "Motores",
    "modulo_abortagem":    "Sistema de abortagem",
}
for chave, nome in modulos.items():
    checar(telemetria[chave] == 1,
           f"{nome}: {'ONLINE' if telemetria[chave] == 1 else 'OFFLINE'} (esperado: ONLINE)")
for a in alertas[n:]: print(a)
```

## 🚦 Célula 3 — Resultado Final

```python
# ============================================================
# RESULTADO FINAL
# ============================================================

total      = len(alertas)
aprovados  = sum(1 for a in alertas if "✅" in a)
reprovados = total - aprovados

print("\n" + "=" * 50)
print(f"  Verificações: {aprovados} aprovadas | {reprovados} com falha")
print("=" * 50)

if aprovado:
    print("\n  🟢 RESULTADO: PRONTO PARA DECOLAR")
    print("     Todos os sistemas estão dentro das faixas seguras.")
else:
    print("\n  🔴 RESULTADO: DECOLAGEM ABORTADA")
    print("     Corrija os itens com ❌ antes de prosseguir.")

print("=" * 50)
```
